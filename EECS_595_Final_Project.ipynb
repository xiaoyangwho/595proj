```python
import nltk
from nltk.corpus import wordnet as wn
nltk.download('wordnet')
```

```python
def summarise(word, v=True, disp=False):
    ret = ''
    ss = wn.synsets(word, pos='v')
    if not v:
        ss = wn.synsets(word)
    for sense in ss:
        ret += 'Sense: %s\n'%(sense.name())
        ret += 'Definition: %s\n'%(sense.definition())
        ret += 'Examples:\n'
        for example in sense.examples():
            ret += '- %s\n'%(example)
        ret += 'Synset:\n'
        for lemma in sense.lemmas():
            frame = ' | '.join(lemma.frame_strings())
            if frame == '':
                frame = 'N/A'
            ret += '- %s (%s)\n'%(lemma.name(), frame)
        ret += '\n'
    if disp:
        print(ret)
    return ret
```

```python
# example usage
k = summarise('peel', v=True, disp=True)
```

```python
def extract_line(line):
    l = len(line)
    w = line[0]
    v = np.zeros(l-1)
    for i in range(1, l):
        v[i-1] = line[i]
    return w, v
```

```python
def extract(filename):
    w2v, i2v, i2w = {}, [], []
    with open(filename) as f:
        for line in f.readlines():
            line = line.split()
            w, v = extract_line(line)
            w2v[w] = v
            i2v.append(v)
            i2w.append(w)
        d = len(i2v[0])
        n = len(i2v)
    return d, n, w2v, i2v, i2w
```

```python
d, n, w2v, i2v, i2w = extract(filename)
```

```python
from scipy.spatial import distance
import numpy as np

def k_nearest(word, k):
    v = w2v[word]
    distances = []
    for i in range(n):
        distances.append(distance.cosine(v, i2v[i]))
    np.argsort(distances)
    sorted_indices = np.argsort(distances)
    k_nearest = []
    for i in range(k):
        k_nearest.append(i2w[sorted_indices[i]])
    print(k_nearest)
    return k_nearest
```

```python
def count(answers):
    yes, no, other = 0, 0, 0
    for i in range(len(answers)):
        answer = answers[i][0]['generated_text'].split('A: ')
        if len(answer) > 1:
            if len(answer[1]) > 2 and (answer[1][0:3] == 'yes' or answer[1][0:3] == 'Yes'):
                yes += 1
            elif len(answer[1]) > 2 and (answer[1][0:2] == 'no' or answer[1][0:2] == 'No'):
                no += 1
            else:
                other += 1
    return yes, no, other
```

So the prior yes:no ratio (without context) is 165:108
